```python
from pylab import *
from scipy.integrate import *
%matplotlib inline
```

```python
def osc(y_v,t,omega2=1.,mu = 0):
    #y_v  = [y,v]
    dy = y_v[1]
    dv = -omega2*y_v[0]- mu*y_v[1]
    return array([dy,dv])
```

```python
inicio = [0,1]
t = linspace(0,100,500)
sol = odeint(osc,inicio,t,args = (1.,0.1))
xlim([-1,1]),ylim([-1,1])
plot(sol[:,0],sol[:,1])
```

```python
from JSAnimation import IPython_display
from matplotlib import animation
```

```python
figura = figure() #Crea la figura
ejes = axes(xlim=(-1,1),ylim=(-1,1))


trayectoria, = ejes.plot([],[],lw=2)
punto, = ejes.plot([],[],'o')

def init():
    trayectoria.set_data([],[])
    punto.set_data([],[])
    return trayectoria, punto,  #Inicializa variables

def el_movimiento(i):
    y,v = sol[:,0],sol[:,1]
    trayectoria.set_data(y[:i+1],v[:i+1])
    punto.set_data(y[i],v[i])
    tache =ejes.plot([0],[0],'x')  #El tache siempre aparece porque tu funcion init no lo contempla
    return trayectoria,punto
    
animation.FuncAnimation(figura,el_movimiento, init_func = init , frames =len(sol[:,0])) # los argumentos son el entorno, lo que va a animar la función borron y cuenta nueva
```
